```python
import json
import pandas as pd

position = pd.read_csv("position.csv")
records = {}
for i in range(0, position.shape[0]):
    records[str(position.iloc[i, 6])] = {"order_id": [-1], "open_qty": int(position.iloc[i, 10]), "close_qty": 0}
with open('DataGT.json', 'w') as jfile:
    json.dump(records, jfile, indent=4)

# 再次读取文件进行检查
with open('DataGT.json', 'r') as jfile:
    gridline_records = json.load(jfile)
total_open_qty, total_close_qty = 0, 0
for gridline, gridinfo in gridline_records.items():
    total_open_qty += gridinfo["open_qty"]
    total_close_qty += gridinfo["close_qty"]
gridline_records = {int(float(k)): {"order_id": [-1], "open_qty": v["open_qty"], "close_qty": v["close_qty"]} for k, v in gridline_records.items()}

print(f"总开仓数量: {total_open_qty}, 总平仓数量: {total_close_qty}")
gridline_records
```
